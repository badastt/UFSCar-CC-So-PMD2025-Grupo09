```python
import requests
import os
import tarfile
import bibtexparser

download_dir = "arxiv_downloads"
extracted_dir = "arxiv_extracted"
os.makedirs(download_dir, exist_ok=True)
os.makedirs(extracted_dir, exist_ok=True)
```

```python
import arxiv

client = arxiv.Client()

search = arxiv.Search(
    query = "quantum",
    max_results = 1000,
    sort_by = arxiv.SortCriterion.SubmittedDate
)

results = client.results(search)

data = {}

for r in client.results(search):

    id = str.split(r.entry_id, '/')[-1]
    link = r.entry_id
    last_update = r.updated
    published = r.published
    title = r.title
    authors = [str(a) for a in r.authors]
    summary = r.summary
    primary_category = r.primary_category
    categories = r.categories
    #print(r.comment)
    #print(r.journal_ref)
    #print(r.doi)
    data[id] = {'link': link, 
                'last_update': last_update.isoformat(), 
                'published': published.isoformat(), 
                'title': title, 
                'authors': authors, 
                'summary': summary, 
                'primary_category': primary_category, 
                'categories': categories,
                'pre-print_link': f'http://arxiv.org/e-print/{id}'}

    print(f"\nProcessing paper {id}: {title}")

    paper_extract_path = os.path.join(extracted_dir, id)
    
    try:
        local_tar_path = os.path.join(download_dir, f"{id}.tar.gz")
        response = requests.get(data[id]['pre-print_link'], stream=True)
        response.raise_for_status()

        with open(local_tar_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        
        paper_extract_path = os.path.join(extracted_dir, id)
        os.makedirs(paper_extract_path, exist_ok=True)

        with tarfile.open(local_tar_path, "r:gz") as tar:
            tar.extractall(path=paper_extract_path)

        bib_files = []
        for root, _, files in os.walk(paper_extract_path):
            for file in files:
                if file.endswith('.bib'):
                    bib_files.append(os.path.join(root, file))

        if bib_files:
            for bib_file_path in bib_files:
                try:
                    with open(bib_file_path, encoding='utf-8') as bibtex_file:
                        parser = bibtexparser.bparser.BibTexParser()
                        #parser.customization = bibtexparser.customization.homogenize_latex_encoding
                        bib_database = bibtexparser.load(bibtex_file, parser=parser)
                        data[id]['bib_entries'] = bib_database.entries
                except Exception as e:
                    print(f"{bib_file_path}: {e}")
        """
        else:
            print("-> No .bib files found. Searching for a .bbl file...")
            bbl_file_found = None
            for root, _, files in os.walk(paper_extract_path):
                for file in files:
                    if file.endswith('.bbl'):
                        bbl_file_found = os.path.join(root, file)
                        break # Found one, no need to look further
                if bbl_file_found:
                    break
            
            if bbl_file_found:
                print(f"-> Found .bbl file: {os.path.basename(bbl_file_found)}. Reading content...")
                try:
                    with open(bbl_file_found, 'r', encoding='utf-8') as bblfile:
                        # We can't parse it, so we just read the raw text
                        data[paper_id]['bbl_content'] = bblfile.read()
                except Exception as e:
                    print(f"!! Error reading {bbl_file_found}: {e}")
            else:
                print("-> No .bbl file found either.")
        """
    except Exception as e:
        print(f"{id}: {e}")
        continue
    

#print(data)

import json

with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```
